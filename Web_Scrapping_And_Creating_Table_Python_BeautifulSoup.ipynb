<p>Extract the data of each of the round of PR invites.</p> 
<p>A table of 3 columns have to be formed which will have data asbelow :</p>
<p>'Date of Draw', 'Number of invitations issued', 'CRS score of lowest-ranked candidate'</p>
<p>For example the first row will have the following data in the 3 columns:</p>
<p>'May 29, 2019' , '3,350' , '470'</p>

```python
# Requests will allow you to send HTTP/1.1 requests and also allows you to access the response data of Python.
import requests  
# for regular expressions
import re
```

```python
# Making a request to web page
page = requests.get("https://www.canada.ca/en/immigration-refugees-citizenship/services/immigrate-canada/express-entry/submit-profile/rounds-invitations/results-previous.html")
```

```python
# check the status of page
page.status_code
```

```python
#Python library for pulling data out of HTML and XML files.
from bs4 import BeautifulSoup as bs
soup = bs(page.content, 'html.parser')    # defining the parser to navigate the page content
```

```python
pattern1=re.compile('Number of invitations issued: ([\d,]+)') # search for pattern
list_invitation=[]  # Creating a empty list to put all the entires for "Number of Invitations Issued"
for para in  soup.find_all('p'):
    if (pattern1.search(para.text)!= None):
# group1 will give the number which will be at position 1, "," will be replaced with "" and then convert to int and will be appended to list
        list_invitation.append(int((pattern1.search(para.text).group(1)).replace(',','')))
list_invitation[27]=list_invitation[27]+list_invitation[28] # since the entry at index 27 is in single para, we will add the data of 27 and 28 and put that in 27 and pop 28
list_invitation.pop(28)
len(list_invitation)
```

```python
#appending rest of the entries that are in table
pattern4=re.compile('(\d\d\d|\d,\d\d\d)')
for para in  soup.find_all(class_='table'):
    list_invitation.append(int((pattern4.search(para.text).group(1)).replace(',','')))
len(list_invitation)  
```

```python
#Date List
pattern2=re.compile(r'[a-zA-Z]+.*')
dates=[data.get_text() for data in soup.find_all('h3')][:-1]
list_Date=[]
for item in dates:
    if pattern2.search(item)!= None:
        list_Date.append(pattern2.search(item).group())

len(list_Date)
```

```python
# CRS Score
pattern3=re.compile('CRS score of lowest-ranked candidate invited: (\d+)')
list_CRS=[]
for para in  soup.find_all('p'):
    if (pattern3.search(para.text)!= None):
        list_CRS.append(int((pattern3.search(para.text).group(1)).replace(',','')))
list_CRS[27]=list_CRS[27]+list_CRS[28]
list_CRS.pop(28)
len(list_CRS)
```

```python
for para in  soup.find_all(class_='table'):
    for td in para.find_all('td')[1:]:
        list_CRS.append(int(td.text[0:4]))
len(list_CRS)     
```

```python
#dict_data.clear()
dict_data={'Date of Draw':list_Date,'Number of invitations issued':list_invitation,'CRS score of lowest-ranked candidate':list_CRS}
dict_data
```

```python
import pandas as pd      
```

```python
df=pd.DataFrame(dict_data)
df
```
